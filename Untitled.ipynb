```python
import time
import threading
import google.generativeai as genai
import speech_recognition as sr
from gtts import gTTS
from IPython.display import Audio, display
from io import BytesIO

# Configure Gemini API
GOOGLE_API_KEY = 'AIzaSyD7sJ2MegXGXdRH4KpiBb6K5JcPrLBb5Js'
genai.configure(api_key=GOOGLE_API_KEY)
model = genai.GenerativeModel('gemini-pro')

def audio_to_text():
    """
    Record audio from the microphone and convert it to text.
    
    Returns:
        str: The text recognized from the audio.
    """
    recognizer = sr.Recognizer()
    with sr.Microphone() as source:
        print("Listening...")
        try:
            audio = recognizer.listen(source, timeout=10, phrase_time_limit=5)
            text = recognizer.recognize_google(audio, show_all=False)
            print(f"Recognized Text: {text}")
            return text
        except sr.UnknownValueError:
            print("Google Speech Recognition could not understand the audio")
            return None
        except sr.RequestError:
            print("Could not request results from Google Speech Recognition service")
            return None

def chat_with_gemini(prompt):
    """
    Send text to Gemini LLM and get the response.

    Args:
        prompt: The text prompt to send to Gemini.

    Returns:
        response: The response from Gemini LLM formatted as a voice assistant response.
    """
    try:
        # Prepend a directive to guide the LLM
        directive = "You are a voice assistant. Please respond as a voice assistant would."
        prompt_with_directive = f"{directive} {prompt}"

        # Generate response from Gemini LLM
        response = model.generate_content(prompt_with_directive)
        response_text = response.text.strip()

        formatted_response = response_text
        return formatted_response
    except Exception as e:
        return f"I encountered an error: {e}"


def text_to_speech(text, audio_file_path):
    """
    Convert text to speech and save it to an audio file.
    
    Args:
        text: The text to convert to speech.
        audio_file_path: Path where the audio file will be saved.
    """
    tts = gTTS(text=text, lang='en')
    tts.save(audio_file_path)
    print("Response converted to speech")

def play_audio(audio_file_path):
    """
    Play the audio file.
    
    Args:
        audio_file_path: Path to the audio file.
    """
    with open(audio_file_path, "rb") as audio_file:
        audio_data = audio_file.read()
        display(Audio(audio_data, autoplay=True))

def process_input():
    """
    Process the audio input and get response from Gemini.
    """
    prompt = audio_to_text()
    if prompt:
        response = chat_with_gemini(prompt)
        print("Gemini: ", response)
        
        audio_file_path = "response.mp3"
        text_to_speech(response, audio_file_path)
        
        # Play the audio response in a separate thread
        playback_thread = threading.Thread(target=play_audio, args=(audio_file_path,))
        playback_thread.start()

def main():
    print('\nReady to chat...')
    while True:
        # Use a separate thread to handle input processing
        input_thread = threading.Thread(target=process_input)
        input_thread.start()
        
        # Wait for the input thread to finish
        input_thread.join()
        
        # Check for exit condition
        exit_prompt = audio_to_text()
        if exit_prompt and exit_prompt.lower() in ['exit', 'quit', 'bye']:
            print("Exiting chat...")
            break
        
        # Sleep briefly to prevent multiple simultaneous recordings
        time.sleep(1)

if __name__ == "__main__":
    main()
```

```python
import os
import google.generativeai as genai
import speech_recognition as sr
from gtts import gTTS
from IPython.display import Audio, display
from io import BytesIO
import time
import playsound
import threading

# Configure the API key
GOOGLE_API_KEY = "AIzaSyD7sJ2MegXGXdRH4KpiBb6K5JcPrLBb5Js"
genai.configure(api_key=GOOGLE_API_KEY)

# Initialize the model
model = genai.GenerativeModel('gemini-pro')

def audio_to_text():
    """
    Record audio from the microphone and convert it to text.
    
    Returns:
        str: The text recognized from the audio.
    """
    recognizer = sr.Recognizer()
    with sr.Microphone() as source:
        print("Listening...")
        try:
            audio = recognizer.listen(source, timeout=3)  # Timeout set to 3 seconds
            text = recognizer.recognize_google(audio)  # No specific language code, detect in default language
            print(f"Recognized Text: {text}")
            return text
        except sr.WaitTimeoutError:
            print("Listening timed out")
            return None
        except sr.UnknownValueError:
            print("Google Speech Recognition could not understand the audio")
            return None
        except sr.RequestError:
            print("Could not request results from Google Speech Recognition service")
            return None

def chat_with_gemini(prompt):
    """
    Send text to Gemini LLM and get the response.
    
    Args:
        prompt: The text prompt to send to Gemini.
    
    Returns:
        response: The response from Gemini LLM.
    """
    try:
        # Prepend a directive to guide the LLM to act as a voice assistant
        directive = "You are a voice assistant. Act like it and just give output as a voice assistant."
        prompt_with_directive = f"{directive} {prompt}"

        start_time = time.time()
        response = model.generate_content(prompt_with_directive)
        result = ''.join([p.text for p in response.candidates[0].content.parts])
        
        # Ensure the response time is within the 3-second limit
        elapsed_time = time.time() - start_time
        if elapsed_time > 3:
            print("Warning: Response time exceeded 3 seconds")
        
        return result
    except Exception as e:
        return f"An error occurred: {e}"

def text_to_speech(text, audio_file_path):
    """
    Convert text to speech and save it to a file.
    
    Args:
        text: The text to convert to speech.
        audio_file_path: Path where the audio file will be saved.
    """
    start_time = time.time()
    tts = gTTS(text=text, lang='en')
    tts.save(audio_file_path)
    
    # Ensure the text-to-speech time is within the 3-second limit
    elapsed_time = time.time() - start_time
    if elapsed_time > 3:
        print("Warning: Text-to-Speech conversion exceeded 3 seconds")

def play_audio(audio_file_path):
    """
    Play the audio file.
    
    Args:
        audio_file_path: Path to the audio file.
    """
    print("Playing response...")
    playsound.playsound(audio_file_path)

def main():
    print('\nReady to chat...')
    while True:
        # Record and convert audio to text
        prompt = audio_to_text()
        if prompt:
            # Send text to Gemini and get the response
            response = chat_with_gemini(prompt)
            print("Gemini: ", response)
            
            # Convert Gemini response to audio and save it
            audio_file_path = "response.mp3"
            text_to_speech(response, audio_file_path)
            
            # Play the audio response
            play_audio(audio_file_path)
            
        # Exit condition
        if prompt and prompt.lower() in ['exit', 'quit', 'bye']:
            print("Exiting chat...")
            break

if __name__ == "__main__":
    main()
```

```python
import tkinter as tk
from tkinter import messagebox
import threading
import google.generativeai as genai
import speech_recognition as sr
from gtts import gTTS
import time
import playsound

# Configure the API key
GOOGLE_API_KEY = "YOUR_GOOGLE_API_KEY"
genai.configure(api_key=GOOGLE_API_KEY)

# Initialize the model
model = genai.GenerativeModel('gemini-pro')

class VoiceAssistantApp:
    def __init__(self, root):
        self.root = root
        self.root.title("Voice Assistant")
        self.root.geometry("400x300")

        # Create interface elements
        self.label = tk.Label(root, text="Press 'Start' to begin", font=("Helvetica", 14))
        self.label.pack(pady=20)

        self.start_button = tk.Button(root, text="Start Listening", command=self.start_listening)
        self.start_button.pack(pady=10)

        self.stop_button = tk.Button(root, text="Stop Listening", state=tk.DISABLED, command=self.stop_listening)
        self.stop_button.pack(pady=10)

        self.text_display = tk.Text(root, height=6, width=40)
        self.text_display.pack(pady=10)

        self.is_listening = False

    def audio_to_text(self):
        """
        Record audio from the microphone and convert it to text.
        """
        recognizer = sr.Recognizer()
        with sr.Microphone() as source:
            self.label.config(text="Listening...")
            try:
                audio = recognizer.listen(source, timeout=3)
                text = recognizer.recognize_google(audio)
                return text
            except sr.WaitTimeoutError:
                return "Listening timed out."
            except sr.UnknownValueError:
                return "Could not understand the audio."
            except sr.RequestError:
                return "Could not request results."

    def chat_with_gemini(self, prompt):
        """
        Send text to Gemini LLM and get the response.
        """
        directive = "You are a voice assistant. Act like it and just give output as a voice assistant."
        prompt_with_directive = f"{directive} {prompt}"

        try:
            response = model.generate_content(prompt_with_directive)
            result = ''.join([p.text for p in response.candidates[0].content.parts])
            return result
        except Exception as e:
            return f"An error occurred: {e}"

    def text_to_speech(self, text):
        """
        Convert text to speech and play it.
        """
        tts = gTTS(text=text, lang='en')
        audio_file_path = "response.mp3"
        tts.save(audio_file_path)
        playsound.playsound(audio_file_path)

    def start_listening(self):
        self.is_listening = True
        self.start_button.config(state=tk.DISABLED)
        self.stop_button.config(state=tk.NORMAL)

        self.listen_thread = threading.Thread(target=self.process_input)
        self.listen_thread.start()

    def stop_listening(self):
        self.is_listening = False
        self.start_button.config(state=tk.NORMAL)
        self.stop_button.config(state=tk.DISABLED)

    def process_input(self):
        while self.is_listening:
            prompt = self.audio_to_text()
            self.text_display.insert(tk.END, f"User: {prompt}\n")
            if prompt:
                response = self.chat_with_gemini(prompt)
                self.text_display.insert(tk.END, f"Assistant: {response}\n")
                self.text_to_speech(response)
            time.sleep(1)  # Prevent the loop from running too fast

        self.label.config(text="Press 'Start' to begin")

def main():
    root = tk.Tk()
    app = VoiceAssistantApp(root)
    root.mainloop()

if __name__ == "__main__":
    main()
```
